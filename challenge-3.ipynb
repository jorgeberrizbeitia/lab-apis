# 1. Read your exported json file from the data folder

```python
import json

with open("data/coffe_shops.json", "r") as file:
    data = json.load(file)
```

# 2. Plot a map with the coffee shops

```python
# import geopandas as gdp
# from shapely.geometry import Point
# from cartoframes.viz import Map, Layer, popup_element

# using matplotlib as I couldn't install geopandas
import matplotlib.pyplot as plt
```

```python
# Extract lat/lon and labels
lats = [item['lat'] for item in data]
lons = [item['long'] for item in data]
names = [item['name'] for item in data]

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(lons, lats, c='red', s=50)

# Add labels
for name, lon, lat in zip(names, lons, lats):
    plt.text(lon + 0.01, lat + 0.01, name, fontsize=9)

plt.title("Locations")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.grid(True)
plt.show()
```
